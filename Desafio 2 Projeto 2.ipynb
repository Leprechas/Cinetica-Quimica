# Desafio 2: Projeto 2

Após diversos problemas com o modelo da simulação atomística 2D, resolvemos trocar para o projeto 2, que consistia em usar dados experimentais, contudo, após uma conversa com o professor responsável no dia 17/10/2023, foi confirmado que poderíamos utilizar os dados iniciais do desafio 1, como o número de partículas/concentração de H2 e O2, e seguir com a situação teórica.

```python
import numpy as np
import matplotlib.pyplot as plt

# Concentrações iniciais
H2_0 = 50
O2_0 = 50
H2O_0 = 0
O_0 = 0

# Constante de velocidade
k = 0.001

# Tempo inicial e final da simulação
t0 = 0
tf = 1

# Número de pontos na simulação
n = 1000

# Intervalo de tempo
dt = (tf - t0) / n

# Vetores para armazenar as concentrações e o tempo
H2 = np.zeros(n)
O2 = np.zeros(n)
H2O = np.zeros(n)
O = np.zeros(n)
t = np.zeros(n)

# Condições iniciais
H2[0] = H2_0
O2[0] = O2_0
H2O[0] = H2O_0
O[0] = O_0
t[0] = t0

# Loop para simular a reação
for i in range(1, n):
    # Cálculo das concentrações no instante i
    v = k * H2[i-1]**2 * O2[i-1]
    H2[i] = H2[i-1] - 2 * v * dt
    O2[i] = O2[i-1] - v * dt
    H2O[i] = H2O[i-1] + v * dt
    O[i] = O[i-1] + v * dt
    t[i] = t[i-1] + dt

# Plot do gráfico
plt.title('Gráfico da []xt, com k = 0.001')
plt.plot(t, H2, label='H2')
plt.plot(t, O2, label='O2')
plt.plot(t, H2O, label='H2O')
plt.plot(t, O, label='O', linestyle='dashed')
plt.xlabel('Tempo')
plt.ylabel('Concentração')
plt.legend()
plt.show()
```
